## 0.0 Solution Planning (IOT)

### 0.1 Input

1. Problema de Negócio.
    - Selecionar os clientes mais valiosos para integrar um programa de fidelização.
    
    
2. Conjunto de Dados.
    - Vendas de um e-commerce online, durante o período de 1 ano.

### 0.2 Output

1. A identificação das pessoas que farão parte do programa de Insiders.
    - Lista: client_id | is_insiders
               264993  |   yes/1
               541556  |   no/0
                
2. Um relatório com as respostas das perguntas de negócio.
    - Quem são as pessoas elegíveis para participar do programa de Insiders?
    - Quantos clientes farão parte do grupo?
    - Quais as principais características desses clientes?
    - Qual a porcentagem de contribuição do faturamento, vinda do Insiders?
    - Qual a expectativa de faturamento desse grupo para os próximos meses?
    - Quais as condições para uma pessoa ser elegível ao Insiders?
    - Quais as condições para uma pessoa ser removida do Insiders?
    - Qual a garantia que o programa Insiders é melhor que o restante da base?
    - Quais ações o time de marketing pode realizar para aumentar o faturamento?

### 0.3 Tasks

1. Quem são as pessoas elegíveis para participar do programa de Insiders?
    - O que é ser elegível? O que é ser um cliente de maior "valor"?
    - Faturamento:
        - Alto Ticket médio. (preço médio das compras do cliente)
        - Alto LTV. (valor total gasto do cliente em todas as compras)
        - Baixa Recência. (tempo desde a última compra)
        - Alto Basket size. (número de produtos da cesta de compra)
        - Baixa probabilidade de churn. (quando o cliente irá parar de comprar na empresa)
    
    - Custo:
        - Baixa taxa de devolução.
        
    - Experiência de compra:
        - Média alta das avaliações.
    
    
2. Quantos clientes farão parte do grupo?
    - Número total de clientes.
    - % do grupo insiders.
    - Calcular a porcentagem de clientes que farão parte do grupo.
    
    
3. Quais as principais características desses clientes?
    - Escrever características desses clientes:
        - Idade.
        - Localização.
        - Etc.
        
    - Escrever características do hábito de consumo:
        - Atributos da clusterização.
        
        
4. Qual a porcentagem de contribuição do faturamento, vinda do Insiders?
    - Faturamento total anual da empresa.
    - Faturamento total anual do grupo insiders.
    - Calcular a porcentagem do faturamento do grupo insiders em relação ao total da empresa.
    
    
5. Qual a expectativa de faturamento desse grupo para os próximos meses?
    - LTV do grupo insiders.
    - Análise de Cohort.
    
    
6. Quais as condições para uma pessoa ser elegível ao Insiders?
    - Decidir a periodicidade do algoritmo (1 mês, 3 meses, etc)
    - A pessoa precisa ser similar ou parecido com uma pessoa do grupo.
    
    
7. Quais as condições para uma pessoa ser removida do Insiders?
    - Decidir a periodicidade do algoritmo (1 mês, 3 meses, etc)
    - A pessoa precisa ser desimilar ou não-parecida com uma pessoa do grupo.
    

8. Qual a garantia que o programa Insiders é melhor que o restante da base?
    - Teste A/B.
    - Teste A/B Bayesiano.
    - Teste de hipótese.
    
    
9. Quais ações o time de marketing pode realizar para aumentar o fatura.
    - Desconto.
    - Preferência de compra.
    - Frete.

## 1.0 Solution Benchmark

### 1.1 Desk Research

```python
1. Modelo RFM (Recency, Frequency, Monetary)
    - Modelo simples de segmentação de clientes.
    - Fácil e prático para desenvolver e aplicar, e ainda assim já é bastante eficiente em termos de retorno financeiro.
    - Possibilita fazer diversas análises para cada grupo de clientes, e entender qual o tipo de ação tomar em relação a cada grupo para melhorar o hábito de consumo de cada grupo.
```
